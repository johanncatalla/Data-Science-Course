```python
dataset_invoice = pd.read_excel("assignment-data.xlsx", sheet_name='online-retail')
dataset = dataset_invoice.head(200)
dataset
```

```python
# Task 1: Cleaning Data
print("before cleaning: any negatives?: ", any(dataset['Quantity'] < 0))
```

```python
dataset['Quantity'] = dataset['Quantity'].abs()
dataset['UnitPrice'] = dataset['UnitPrice'].abs()
dataset
```

```python
print("after cleaning: any negatives?: ", any(dataset['Quantity'] < 0))
```

```python
# Task 2: Find out the customer who has done the most number of transactions
customer_transactions = dataset.groupby('CustomerID')['InvoiceNo'].nunique().reset_index()
customer_transactions = customer_transactions.rename(columns={'InvoiceNo': 'TransactionCount'})
customer_with_most_transactions = customer_transactions[customer_transactions['TransactionCount'] == customer_transactions['TransactionCount'].max()]
customer = int(customer_with_most_transactions['CustomerID'].values[0])
number_of_transactions = customer_with_most_transactions['TransactionCount'].values[0]

print(f'Hurray! The most numbr of transactions is done by CustomerID: {int(customer)}. He has made {number_of_transactions} transactions. He will be rewarded with our gift hamper worth 1000 pesos')
```

```python

# Task 3: Find the Customer with the highest amount of total purchase amount.
# Calculate the total purchase amount for each customer
dataset['TotalPurchaseAmount'] = dataset['Quantity'] * dataset['UnitPrice']
customer_total_purchase = dataset.groupby('CustomerID')['TotalPurchaseAmount'].sum().reset_index()

# Find the customer with the highest total purchase amount
customer_with_highest_purchase = customer_total_purchase[customer_total_purchase['TotalPurchaseAmount'] == customer_total_purchase['TotalPurchaseAmount'].max()]
customer_highest_purchase = int(customer_with_highest_purchase['CustomerID'].values[0])
highest_purchase_amount = customer_with_highest_purchase['TotalPurchaseAmount'].values[0]

print(f"Customer ID:{customer_highest_purchase} is the best customer since he has spent a lot on our business. His total purchase amount is {highest_purchase_amount:.2f} pesos. 50% discount will be given on his next purchase.")
```

```python
# Task 4: Amazon wants to find out the most and the least bought items.
# Group by 'StockCode' and sum the 'Quantity' for each stock item
most_and_least = dataset.groupby('StockCode')['Quantity'].sum()

# Convert the result to a DataFrame for better readability
most_and_least_df = most_and_least.reset_index(name='NoOfPurchase')

# Sort the DataFrame by 'NoOfPurchase' in descending order
sorted_df = most_and_least_df.sort_values(by='NoOfPurchase', ascending=False)

# Get the most and least purchased items
most = sorted_df.iloc[0]
least = sorted_df.iloc[-1]

print(f'Most bought item is Stock Code: {most["StockCode"]} with {most["NoOfPurchase"]} in purchased quantity!')
print(f'Least bought item is Stock Code: {least["StockCode"]} with {least["NoOfPurchase"]} in purchased quantity!')
print(f'These two items can be bundled and provide a 10% discount to increase the sale of Stock Code: {least["StockCode"]}.')
```

```python
# TASK 5: OVERALL AVERAGE

average_purchase_amount = dataset['TotalPurchaseAmount'].mean()
print(f"Average purchase amount among all transactions: {average_purchase_amount:.2f}")
```

```python
# TASK 6: find out the total purchase amount and the number of transactions made by each customer in their lifetime  

dataset.loc[:, 'product_total'] = dataset['Quantity'] * dataset['UnitPrice']

grouped = dataset.groupby('CustomerID').agg(
    product_total=('product_total', 'sum'),
    NumberofTransactions=('InvoiceNo', 'count')
)

new_dataset = grouped.reset_index()
new_dataset.tail(10)

```

```python
# Find the average purchase amount made by every customer . 

new_dataset['AveragePurchase'] = new_dataset['product_total'] / new_dataset['NumberofTransactions']
new_dataset.head(5)
```
