Description of each column in the dataset:

1. **Age**: The age of the individual.
2. **Gender**: The gender of the individual (e.g., Male, Female).
3. **Height**: The height of the individual in meters.
4. **Weight**: The weight of the individual in kilograms.
5. **CALC**: Unknown column. You might need to check the data source or documentation to understand what this column represents.
6. **FAVC**: Whether the individual frequently consumes high caloric food (e.g., yes, no).
7. **FCVC**: Frequency of consumption of vegetables (numeric scale).
8. **NCP**: Number of main meals per day (numeric scale).
9. **SCC**: Squamous cell carcinoma
10. **SMOKE**: Whether the individual smokes (e.g., yes, no).
11. **CH2O**: Consumption of water daily (numeric scale).
12. **family_history_with_overweight**: Whether the individual has a family history of overweight (e.g., yes, no).
13. **FAF**: Physical activity frequency (numeric scale).
14. **TUE**: Time using technology devices (numeric scale).
15. **CAEC**: Unknown column. You might need to check the data source or documentation to understand what this column represents.
16. **MTRANS**: Mode of transportation (e.g., Public Transportation, Walking).
17. **NObeyesdad**: Obesity level of the individual (e.g., Normal_Weight, Overweight_Level_I).

Some columns have numeric values, some are categorical, and others seem to be ordinal.

## Question 1 - 3: simple EDA

```python
# Average age of individuals in the dataset?
average_age = data['Age'].mean()
print("Average Age:", average_age)
```

```python
# What is the distribution of weight based on gender?

weight_by_gender = data.groupby('Gender')['Weight'].mean()
print("\nWeight Distribution by Gender:")
print(weight_by_gender)
```

```python
# What is the most common transportation mode?
common_transportation = data['MTRANS'].value_counts().idxmax()
print("\nMost Common Transportation Mode:", common_transportation)
```

```python
# How many individuals have a family history of being overweight?
family_history_yes_count = data[data['family_history_with_overweight'] == 'yes'].shape[0]
print("\nNumber of individuals with a family history of being overweight:", family_history_yes_count)
```

```python
# What is the distribution of obesity levels (NObeyesdad)?
obesity_level_counts = data['NObeyesdad'].value_counts()
print("\nObesity Level Distribution:")
print(obesity_level_counts)
```

```python
# 4. What is the Height and Weight Distribution?
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='Height', color='skyblue', kde=True, label='Height')
plt.title('Height Distribution')
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.legend()
plt.show()

```

```python
# Weight Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='Weight', color='salmon', kde=True, label='Weight')
plt.title('Weight Distribution')
plt.xlabel('Weight')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# 5. Top 10 Ages with Highest Weight
top_10_ages = data.groupby('Age')['Weight'].mean().nlargest(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_ages.index, y=top_10_ages.values, color='skyblue')
plt.title('Top 10 Ages with Highest Weight')
plt.xlabel('Age')
plt.ylabel('Average Weight')
plt.xticks(rotation=45)
plt.show()
```

```python
# 6. Average Ages with Family History with Overweight
plt.figure(figsize=(8, 5))
sns.boxplot(data=data, x='family_history_with_overweight', y='Age', palette='Set2')
plt.title('Distribution of Ages by Family History of Overweight')
plt.xlabel('Family History with Overweight')
plt.ylabel('Age')
plt.show()
```

```python
# Violin plot of Average Ages with Family History with Overweight
plt.figure(figsize=(10, 6))
sns.violinplot(data=data, x='family_history_with_overweight', y='Age', palette='Set2')
plt.title('Distribution of Ages by Family History of Overweight')
plt.xlabel('Family History with Overweight')
plt.ylabel('Age')
plt.show()
```

```python
# 7. Calculate the correlation matrix
# Select only numeric columns
numeric_df = data.select_dtypes(include=['number'])

# Calculate correlation matrix
correlation_matrix = numeric_df.corr()


# Print the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# 8. What is the relationship between smoking and obesity levels?
smoke_obesity_relation = data.groupby('SMOKE')['NObeyesdad'].value_counts()
print("\nSmoking and Obesity Levels:")
print(smoke_obesity_relation)
```

```python
# 9. Do males or females tend to consume more vegetables (FCVC)?
avg_vegetable_consumption_by_gender = data.groupby('Gender')['FCVC'].mean()
print("\nAverage Vegetable Consumption by Gender:")
print(avg_vegetable_consumption_by_gender)
```

Females tend to consume more vegetables.

```python
# 10. What is the height distribution across different obesity levels?
plt.figure(figsize=(8, 6))
data.boxplot(column='Height', by='NObeyesdad')
plt.title('Height Distribution Across Obesity Levels')
plt.suptitle('')  # Removing the default title
plt.ylabel('Height (meters)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
